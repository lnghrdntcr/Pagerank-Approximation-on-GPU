```python
import seaborn as sns
from math import log
import matplotlib.pyplot as plt
sns.set_context("talk")
```

```python
ranges = [i for i in range(1, 20)]
error_values = [log(10**-i, 10) / log(0.85, 10) for i in ranges]
```

```python
r = [10**-i for i in range(1, 20)]
ax = sns.lineplot(x=r, y=error_values)
plt.xlim((max(r), min(r)))
plt.xscale("log")
```
